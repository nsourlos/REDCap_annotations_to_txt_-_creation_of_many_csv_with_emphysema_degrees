```python
#Import libraries
import pandas as pd
import numpy as np
import os

#Create a list paths containing the names of four different paths. 
#Each path represents a different degree of emphysema, as indicated by the names of the paths.
paths=['advanced_gt','confluent_gt_only_new_added','moderate_gt_only_new_added',
       'noemphysema_gt_only_new_added'] #Paths with the csv files of annotations

#path to the folder containing the csv files of annotations.
ground_truth_path="H:\My Desktop/emphysema_exp_scans/" 

path_to_save="H:\My Desktop/" #the path where the output file should be saved.

for path in paths: #Loop over all emphysema degree paths
    
    #Open a txt file to save all annotations for specific degree of emphysema
    with open(path_to_save+'output_'+path+'.txt', 'w') as f: 
        
        for file in os.listdir(ground_truth_path+path): #Loop over csv files of given degree
        
            f.write(str(file)) #Write participant id to txt file
            f.write('\n')

            ground_truth_nodules=pd.read_csv(ground_truth_path+path+'/'+file) #reads the current csv file
            slice_cols=[col for col in ground_truth_nodules.columns if 'slice' in col]
            vals=ground_truth_nodules[slice_cols].values #extract values
            vals=vals[~np.isnan(vals)] #remove nan values
            vals=vals.astype(int) #Convert to integers
            
            f.write("Slices are ") #write slices to txt file
            f.write(str(vals))
            f.write('\n')

            ids=[col for col in ground_truth_nodules.columns if 'nodule_id' in col]
            nodule_ids=ground_truth_nodules[ids].values #extract values
            nodule_ids=nodule_ids[~np.isnan(nodule_ids)] #remove nan values
            nodule_ids=nodule_ids.astype(int) #convert to integers
            
            f.write("Final nod_ids: ") #Write them to txt file
            f.write(str(nodule_ids))
            f.write('\n')

            #similar as above for volumes
            vol_ids=[col for col in ground_truth_nodules.columns if 'volume_solid' in col]
            volume_ids=ground_truth_nodules[vol_ids].values
        
            f.write("nods with solid ")
            f.write(str(volume_ids))
            f.write('\n')
            
            #find where volumes missing and replace with subsolid component
            missing=np.where(np.isnan(volume_ids))
            vol_ids_sub=[col for col in ground_truth_nodules.columns if 'volume_subsolid' in col]
            volume_ids_sub=ground_truth_nodules[vol_ids_sub].values
            volume_ids_sub=volume_ids_sub[missing]
            
            f.write("nods with subsolid ")
            f.write(str(volume_ids_sub))
            f.write('\n')

            volume_ids[np.isnan(volume_ids)]=volume_ids_sub
            volume_ids=volume_ids[~np.isnan(volume_ids)]
            volume_ids=volume_ids.astype(float)
            
            f.write("Final vols: ")
            f.write(str(volume_ids))
            f.write('\n')
            f.write('\n')

        f.close()
```
